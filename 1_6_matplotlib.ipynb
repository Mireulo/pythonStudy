# matplotlib
matplotlib은 그래프를 그려주는 라이브러리다. matplotlib을 사용하면 그래프 그리기와 데이터 시각화가 쉬워진다.

그래프를 그리려면 matplotlib의 pyplot 모듈을 이용해야한다. sin 함수를 그리는 예를 살펴보자

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1)  # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```


## **그래프 생성과정**
1.   넘파이의 arange 메소드로 [0, 0.1, 0.2, ..., 5.8, 5.9]라는 데이터를 생성하여 변수 x에 할당
2.   x의 각 원소에 넘파이의 sin 함수인 np.sin()을 적용하여 변수 y에 할당
3.   x와 y를 인수로 plt.plot 메소드를 호출해 그래프를 그린다
4.   마지막으로 plt.show()를 호출해 그래프를 화면에 출력하고 끝남



```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1)    # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')    # cos 함수는 점선으로 그리기
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.yticks(np.arange(-1.0, 1.5, 0.5))
plt.title('sin & cos')  # 제목
plt.legend()
plt.show()
```
